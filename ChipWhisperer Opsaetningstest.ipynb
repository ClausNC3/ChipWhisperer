Hvis du bruger en af følgende, skal du have `arm-none-eabi-gcc` compileren:

* ChipWhisperer-Lite 32-bit (med ARM-target).
* ChipWhisperer UFO-kort med ARM-target (f.eks. STM32F3, STM32F0 osv.).
* ChipWhisperer-Nano
* ChipWhisperer-Husky med ARM-target (f.eks. SAM4S)

Du kan nemt kontrollere for den fungerende kompiler med følgende, som bør udskrive versionen og kompileroplysninger:

```python
!arm-none-eabi-gcc -v
```

Hvis du bruger en af følgende, skal du have `avr-gcc` kompileren:

* ChipWhisperer-Lite Classic (med XMEGA-target).
